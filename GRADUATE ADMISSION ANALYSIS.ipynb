```python
Y=final_df['Chance of Admit ']
X=final_df.drop('Chance of Admit ',axis=1)
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
dt={'Actual': Y_test, 'Predicted ': y_pred}
data=pd.DataFrame(dt)
data
```

```python
data.to_csv("Linear Regression Results.csv")
```

```python
#Errors 

from sklearn import metrics 

print(metrics.mean_absolute_error(Y_test,y_pred))
print(metrics.mean_squared_error(Y_test,y_pred))
print(np.sqrt(metrics.mean_squared_error(Y_test,y_pred)))
print(metrics.explained_variance_score(Y_test,y_pred))
```

Keeping track of every single metric can get tedious, so we pick one or two metrics to evaluate our model. A good practice is to make sure that the mean squared error is low and the explained variance score is high.

```python
#Accuracy 
print(metrics.r2_score(Y_test,y_pred)*100)
```

An R2 score near 1 means that the model is able to predict the data very well.
